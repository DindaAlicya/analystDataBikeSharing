## Data Wrangling

mari berkenalan dengan data wragling, data wragling merupakan proses membersihkan, mengubah, dan menyusun data mentah agar lebih siap digunakan untuk analisis atau pemodelan. Dalam data wrangling, kita memastikan bahwa data yagn digunakan bebas dari kesalahan, duplikasi, atau nilai yang hilang, serta dalam format yang sesuai untuk diproses lebih lanjut.

Untuk memastikan data tersebut sudah dalam format yang sesuai untuk diproses lebih lanjut, data wrangling memiliki tahapan-tahapan untuk membersihkan, mengubah, dan menyusun data sesuai dengan kebutuhan analisis, yakni:

- gathering data:

proses pengumpulan data. Pada tahap ini kita akan mengumpulkan semua data yang dibutuhkan untuk menjawab semua pertanyaan atau masalah bisnis yang ingin kita hadapi.

- Assesing Data:

proses penilaian data. Pada tahap ini kita akan menilai kualitas dan struktur dari sebuah data. Selain itu, proses ini juga bertujuan untuk mengidentifikasi berbagai masalah yang terdapat dalam data, seperti missing value, unstandard value, dll.

- Cleaning Data:

pada proses Assesing kita menemukan masalah (missing value, outlier, dll.) yang terdapat di dalam sebuah data, masalah tersebut harus dibersihkan sebelum masuk tahap analisis data.

### Cleaning Data

di Cleaning data kita akan menyelesaikan permasalahan yang terdapat pada assesing data yang telah di jabarkan

```python
# mengganti nama kolom agar lebih jelas

print("                        =====Rename Data Day=====")
day_df.rename(columns={
    'dteday' : 'date',
    'instant' : 'instant',
    'season' : 'season',
    'yr' : 'year',
    'mnth' : 'month',
    'holiday' : 'holiday',
    'weekday' : 'weekday',
    'workingday' : 'workingday',
    'weathersit' : 'weather',
    'temp' : 'temperature',
    'hum' : 'humidity',
    'windspeed' : 'windspeed',
    'registered' : 'user_registered',
    'casual' : 'user_casual',
    'cnt' : 'total_rent'
}, inplace=True
)
print(day_df)
print()
print("                      ======Rename Data Hour====")
hour_df.rename(columns={
    'dteday' : 'date',
    'instant' : 'instant',
    'season' : 'season',
    'yr' : 'year',
    'mnth' : 'month',
    'hr' : 'hour',
    'holiday' : 'holiday',
    'weekday' : 'weekday',
    'workingday' : 'workingday',
    'weathersit' : 'weather',
    'temp' : 'temperature',
    'hum' : 'humidity',
    'windspeed' : 'windspeed',
    'registered' : 'user_registered',
    'casual' : 'user_casual',
    'cnt' : 'total_rent'
}, inplace=True
)
print(hour_df)
```

```python
# mengubah tipe data yang belum sesuai

day_df['date'] = pd.to_datetime(day_df['date'])
hour_df['date'] = pd.to_datetime(hour_df['date'])

day_df.info()
hour_df.info()
```

```python
# melakukan mapping tiap Kolom dengan Unique Value (ganti dengan nama yang sesuai)

change = {0 : 2011, 1 : 2012}
day_df.replace({'year': change}, inplace=True)
change = { 1 : 'Jan' , 2 : 'Feb' , 3 : 'Mar' , 4 : 'Apr' , 5 : 'May' , 6 : 'Jun' , 7 : 'Jul' , 8 : 'Aug', 9 : 'Sep' , 10 : 'Oct' , 11 : 'Nov' , 12 : 'Dec' }
day_df.replace({'month': change}, inplace=True)
change = {0 : "Not holiday", 1 : "Holiday"}
day_df.replace({'holiday': change}, inplace=True)
change = {0 : 'Su', 1 : 'Mo',  2 : 'Tu',  3 : 'We',  4 : 'Th', 5 : 'Fr', 6 : 'Sa' }
day_df.replace({'weekday': change}, inplace=True)
change = {0 : "Not Workday", 1 : "Workingday"}
day_df.replace({'workingday': change}, inplace=True)
change = { 1 : 'Spring' , 2 : 'Summer' , 3 : 'Fall' , 4 : 'Winter' }
day_df.replace({'season': change}, inplace=True)
change = { 1 : 'Clear/Few clouds' , 2 : 'Cloudy/Broken clouds' , 3 : 'Light Rain/Snow' , 4 : 'Heavy Rain/Snow' }
day_df.replace({'weather': change}, inplace=True)
day_df

change = {0 : 2011, 1 : 2012}
hour_df.replace({'year': change}, inplace=True)
change = { 1 : 'Jan' , 2 : 'Feb' , 3 : 'Mar' , 4 : 'Apr' , 5 : 'May' , 6 : 'Jun' , 7 : 'Jul' , 8 : 'Aug', 9 : 'Sep' , 10 : 'Oct' , 11 : 'Nov' , 12 : 'Dec' }
hour_df.replace({'month': change}, inplace=True)
change = {0 : "Not holiday", 1 : "Holiday"}
hour_df.replace({'holiday': change}, inplace=True)
change = {0 : 'Sunday', 1 : 'Monday',  2 : 'Tuesday',  3 : 'Wednesday',  4 : 'Thursday', 5 : 'Friday', 6 : 'Saturday' }
hour_df.replace({'weekday': change}, inplace=True)
change = {0 : "Not Workday", 1 : "Workingday"}
hour_df.replace({'workingday': change}, inplace=True)
change = { 1 : 'Spring' , 2 : 'Summer' , 3 : 'Fall' , 4 : 'Winter' }
hour_df.replace({'season': change}, inplace=True)
change = { 1 : 'Clear/Few clouds' , 2 : 'Cloudy/Broken clouds' , 3 : 'Light Rain/Snow' , 4 : 'Heavy Rain/Snow' }
hour_df.replace({'weather': change}, inplace=True)
hour_df
```

```python
# penyelesaian data yang terdapat di outlier
print("==========penyelesaian outlier dalam data Day===========")
Q1 = day_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']].quantile(0.25)
Q3 = day_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']].quantile(0.75)
IQR = Q3 - Q1

max_val = Q3 + (1.5 * IQR)
min_val = Q1 - (1.5 * IQR)

under = (day_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']] < min_val).any(axis=1)
upper = (day_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']] > max_val).any(axis=1)


day_df.drop(day_df[under].index, inplace=True)
day_df.drop(day_df[upper].index, inplace=True)
day_df.info()
print()

print("==========penyelesaian outlier dalam data Hour===========")
Q1 = hour_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']].quantile(0.25)
Q3 = hour_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']].quantile(0.75)
IQR = Q3 - Q1

max_val = Q3 + (1.5 * IQR)
min_val = Q1 - (1.5 * IQR)

under = (hour_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']] < min_val).any(axis=1)
upper = (hour_df[['windspeed', 'user_casual', 'humidity', 'user_registered', 'total_rent']] > max_val).any(axis=1)

hour_df.drop(hour_df[under].index, inplace=True)
hour_df.drop(hour_df[upper].index, inplace=True)
hour_df.info()
```

**Insight:**
- mengubah nama kolom agar lebih jelas
- mengubah tipe data yang belum sesuai
- melakukan mapping pada unique value
- membenahi data yang terdapat di outlier (melakukan dropping)

tahap data wragling pun selesai dan akan dilanjutkan exploratory Data Analysis

### Pertanyaan 1:
**Musim apa yang memiliki jumlah penyewaan tertinggi dan terendah? lalu bagaimana cara meningkatkan penyewaan pada musim dengan jumlah penyewaan rendah?**

```python
# berikut visualisasi nya


result = day_df.groupby(by="season").agg({
    "user_registered": "sum",
    "user_casual": "sum",
    "total_rent":  "sum"
}).sort_values(by="total_rent", ascending=False)

plt.figure(figsize=(7, 4))
sns.barplot(data=result.reset_index(), x="season", y="total_rent", hue="season", palette="inferno")
for index, value in enumerate(result["total_rent"]):
    plt.text(index, value + 100, f"{int(value)}", ha='center', fontsize=8)
plt.title('Total Penyewaan Sepeda Berdasarkan musim (Day)')
plt.xlabel('musim')
plt.ylabel('Total Penyewaan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()



result = hour_df.groupby(by="season").agg({
    "user_registered": "sum",
    "user_casual": "sum",
    "total_rent":  "sum"
}).sort_values(by="total_rent", ascending=False)

plt.figure(figsize=(7, 4))
sns.barplot(data=result.reset_index(), x="season", y="total_rent", hue="season", palette="inferno")
for index, value in enumerate(result["total_rent"]):
    plt.text(index, value + 100, f"{int(value)}", ha='center', fontsize=8)
plt.title('Total Penyewaan Sepeda Berdasarkan musim (Hour)')
plt.xlabel('musim')
plt.ylabel('Total Penyewaan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

dari diagram diatas kita bisa menyimpulkan bahwa:
- penyewaan sepeda terendah berada pada musim **spring**
- penyewaan sepeda tertinggi berada pada musim **fall**

kita bisa meningkatkan penyewaan pada musim spring dengan melakukan promosi pada musim tersebut seperti:
- memberi paket berkelompok, ex: harga akan lebih murah jika menyewa dengan 4 sepeda sekaligus
- memberikan reward limited edition pada penyewa musim tersebut contohnya seperti hiasan meja yang bertema musim spring

### Pertanyaan 3:


**Bagaimana cara meningkatkan penyewaan sepeda di waktu saat penyewaan sepeda cenderung rendah?:**

```python

day_df['weekday'] = day_df['weekday'].astype(str)
weekday_rentals = day_df.groupby('weekday')['total_rent'].sum().reset_index().sort_values(by="total_rent", ascending=True)

plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_rentals['weekday'], y=weekday_rentals['total_rent'], hue=weekday_rentals['weekday'], palette='magma')
for index, value in enumerate(weekday_rentals["total_rent"]):
    plt.text(index, value + 100, f"{int(value)}", ha='center', fontsize=10)
plt.title("Peminjaman Sepeda Berdasarkan Hari (Weekday)")
plt.xlabel("Hari")
plt.ylabel("Total Penyewaan")
plt.show()
```

```python
day_df['workingday'] = day_df['workingday'].astype(str)
workingday_rentals = day_df.groupby('workingday')['total_rent'].sum().reset_index().sort_values(by="total_rent", ascending=True)


plt.figure(figsize=(6, 5))
sns.barplot(x=workingday_rentals['workingday'], y=workingday_rentals['total_rent'], hue=workingday_rentals['workingday'], palette='magma')
for index, value in enumerate(workingday_rentals["total_rent"]):
    plt.text(index, value + 100, f"{int(value)}", ha='center', fontsize=10)
plt.title("Peminjaman Sepeda pada Working vs Non-Working Day")
plt.xlabel("Hari")
plt.ylabel("Total Penyewaan")
plt.show()
```

```python
day_df['holiday'] = day_df['holiday'].astype(str)
holiday_rentals = day_df.groupby('holiday')['total_rent'].sum().reset_index().sort_values(by="total_rent", ascending=True)


plt.figure(figsize=(6, 5))
sns.barplot(x=holiday_rentals['holiday'], y=holiday_rentals['total_rent'], hue=holiday_rentals['holiday'], palette='magma')
for index, value in enumerate(holiday_rentals["total_rent"]):
    plt.text(index, value + 100, f"{int(value)}", ha='center', fontsize=10)
plt.title("Peminjaman Sepeda pada holiday vs Non-holiday")
plt.xlabel("Hari")
plt.ylabel("Total Penyewaan")
plt.show()
```

```python
hour_rentals = hour_df.groupby('hour')['total_rent'].sum().reset_index().sort_values(by="total_rent", ascending=True)

plt.figure(figsize=(10, 5))
sns.lineplot(x=hour_rentals['hour'], y=hour_rentals['total_rent'], marker='o', color='red')
plt.xticks(ticks=range(0, 24), labels=[f"{i:02d}:00" for i in range(24)])
plt.title("Peminjaman Sepeda Berdasarkan Jam")
plt.xlabel("Jam")
plt.ylabel("Total Penyewaan")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

dari data tersebut, menunjukkan bahwa waktu penyewaan yang cenderung rendah adalah saat:
- hari weekend (sabtu & minggu)
- not working day (bukan hari kerja)
- pada dini hari (jam 23:00-4:00)
